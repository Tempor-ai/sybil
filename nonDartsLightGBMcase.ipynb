```python
import lightgbm as lgb
from sklearn.datasets import load_diabetes
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Load the diabetes dataset
diabetes = load_diabetes()
X = diabetes.data
y = diabetes.target

# Create a LightGBM dataset
train_data = lgb.Dataset(X, label=y)

# Set hyperparameters with verbosity control
params = {
    'objective': 'regression',
    'metric': 'l2',
    'boosting_type': 'gbdt',
    'verbose': 10,  # Set to -1 to turn off verbose output
}
num_round = 100

# Train the model
bst = lgb.train(params, train_data, num_round)

# Extract training results
evals_result = bst.eval_train()

# Extract training MSE
train_mse = evals_result['l2']
x_axis = range(1, len(train_mse) + 1)

# Plot the training curve
plt.figure(figsize=(8, 6))
plt.plot(x_axis, train_mse, label='Train MSE')
plt.xlabel('Number of Rounds')
plt.ylabel('Mean Squared Error')
plt.title('LightGBM Training Curve (Diabetes Dataset)')
plt.legend()
plt.show()
```
